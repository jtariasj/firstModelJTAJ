```python
import seaborn as sb
sb.scatterplot(x = "celsius", y ="fahrenheit", data = datos, hue="fahrenheit", palette="coolwarm")
```

```python

#entradas (X) salida (y)
X=datos["celsius"]
y=datos["fahrenheit"]
#cada dato (X,y) es un arreglo de los datos
#celsius = [-22,-4, 14,......]
#y queremos que pase a un arreglo de arreglos de la siguiente manera
# celsius = [[-22],[-4],[14],......]
transformedX = X.values.reshape(-1,1)
transformedy= y.values.reshape(-1,1)
type(datos)
```

```python
type(X)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Entrenamiento
model.fit(transformedX, transformedy)
```

```python
#Predicciones
model.predict([[108]])
```

```python
model.score (transformedX, transformedy)
```
